```python
#list of numbers
numbers=[10,5,30,40,1,60]
even_count=0
odd_count=0
for num in numbers:
    if num%2==0:
        even_count+=1
    else:
        odd_count+=1
print("EVEN Numbers:", even_count)
print("ODD Numbers:",odd_count)
```
